```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


prices = [
	(1, 2.12),
	(2, 2.56),
	(3, 3.10),
	(4, 3.16),
	(5, 3.58),
	(6, 5.12),
	(7, 5.16),
	(8, 5.20),
	(9, 4.12),
	(10, 4.10),
	(11, 3.65),
	(12, 4.25),
]


df = pd.DataFrame(prices, columns=["month", "prices"])
df = df.set_index("month")
plt.title("Price of goods (USD)")
plt.xlabel("Month")
plt.ylabel("Prices")
plt.xlim([0.5,12.5])
plt.ylim([0,2])

prices_in_USD = df['prices'].apply(lambda prices: prices * 1/4)
df['prices'] = prices_in_USD

plt.grid(True, linewidth=0.5, color='#ff0000', linestyle='-')
plt.plot(df.index, df["prices"], 'r--')
```
